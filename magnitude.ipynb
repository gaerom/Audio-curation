```python
import numpy as np
import os
import librosa
import librosa.display
```

```python
# audio magnitude의 최댓값만 따로 저장
input_file_path = 'mean_median_max_16.txt'


def extract_max_values(input_file_path):
    max_values = []
    with open(input_file_path, "r") as file:
        for line in file:
            parts = line.split(': ')
            
            file_name = parts[0]
            
            max_value = line.split("Max: ")[1].split(" /")[0].strip()
            max_values.append(f"{file_name}: {max_value}")
    
    output_file_path = '/mnt/storage1/vggsoundsync/extracted_max_values_16.txt'
    with open(output_file_path, "w") as file:
        for max_value in max_values:
            file.write(max_value + "\n")
    
    return output_file_path


output_file_path = extract_max_values(input_file_path)
```

```python
# 위에서 max만 따로 저장한 txt 읽어와서 값이 0인 것의 개수 파악
output_file = 'extracted_max_values_16.txt'

with open(output_file, 'r') as f:
    lines = f.readlines()

zero = []
for line in lines:
    filename, value = line.split(':')
    if value.strip() == '0.0':
        zero.append(filename)


for i in zero:
    print(i)

print(f'magnitude 최댓값이 0인 audio 개수: {len(zero)}') # 803 / 162,031

zero_file = 'zero_16.txt'
with open(zero_file, "w") as file:
    for j in zero:
        file.write(j + "\n")
```

```python
# 위에서 알아낸 0인 파일들만 다른 dir로 이동시키기 (완전 삭제는 X)
import shutil
import os

source_dir = '/mnt/storage1/trainvideo_10_audios'
target_dir = 'zero_audios'

if not os.path.exists(target_dir):
    os.makedirs(target_dir)
 
with open('zero_16.txt', 'r') as file:
    for line in file:
        filename = line.strip()

        source_file = os.path.join(source_dir, filename)
        target_file = os.path.join(target_dir, filename)
        
        if os.path.exists(source_file):
            shutil.move(source_file, target_file) # 803개 이동
```

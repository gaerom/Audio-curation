```python
import numpy as np
import os
import librosa
import librosa.display
```

```python
output_file = 'max_amplitude_values.txt'

with open(output_file, 'r') as f:
    lines = f.readlines()

zero = []
for line in lines:
    filename, value = line.split(':')
    if value.strip() == '0.0':
        zero.append(filename)


for i in zero:
    print(i)

print(f'magnitude 최댓값이 0인 audio 개수: {len(zero)}') # 512 / 103,797 
```
